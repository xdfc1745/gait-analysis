```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import csv

from pandas import DataFrame, Series
from sklearn.preprocessing import MinMaxScaler

%matplotlib inline
```

```python
def print_graph(MinMax, num): # num is cut range
    length = len(MinMax)
    plt.figure(figsize=(20, 5)) # graph size
    plt.xticks(np.arange(0, length+1, num))
    plt.plot(MinMax, 'ro')
    plt.show()
```

```python
def cut_count(MinMax, num, cnt): # 왼발, 오른발의 데이터를 잘라서 구분
    count = int((len(MinMax)-num) / cnt)
    step = []
    temp = 0
    for i in range(0, count):
        temp = num+i*cnt
        step.append(MinMax[temp:temp+cnt])
    return step
```

```python
rawData = pd.read_csv('normal1/normalWalk0.csv')
xyz = rawData[['AccX', 'AccY', 'AccZ']]
print_graph(xyz, 200)
x = rawData['AccX']
y = rawData['AccY']
z = rawData['AccZ']
svm = abs(x) + abs(y) + abs(z)
print_graph(svm, 200)
step = cut_count(svm, 0, 200)
print_graph(step, 200)
len(step)
for i in step :
    print(max(i))
    print_graph()
```
