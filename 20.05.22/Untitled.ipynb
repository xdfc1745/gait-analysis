```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import csv

from pandas import DataFrame, Series
from sklearn.preprocessing import MinMaxScaler

%matplotlib inline
```

```python
def print_graph(MinMax, num): # num is cut range
    length = len(MinMax)
    plt.figure(figsize=(20, 5)) # graph size
    plt.xticks(np.arange(0, length+1, num))
    plt.plot(MinMax, 'ro')
    plt.show()
```

```python
for i in range(0, 5):
    rawData = pd.read_csv(f'YooHyungjin/normalWalk{i}.csv')
#     print(len(rawData))
    columns = rawData['AccX']
    print_graph(columns, 100)
#     print(result)
```
